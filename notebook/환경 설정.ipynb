#### BERT multilingual model

[BERT](https://github.com/google-research/bert/blob/master/multilingual.md)

- BERT-Base, Multilingual Cased (New, recommended): 104 languages, 12-layer, 768-hidden, 12-heads, 110M parameters

```
## BERT 
import urllib.request
import zipfile

SAVE_DIR = '/content/gdrive/My Drive/ai-summit/model/'

def download(url, savedir=SAVE_DIR, zipped=True):
  file_name = url.split('/')[-1]
  download_file = savedir + file_name
  urllib.request.urlretrieve(url, download_file)

  if zipped:
    zip = zipfile.ZipFile(download_file)
    zip.extractall(savedir)
    zip.close()

url = 'https://storage.googleapis.com/bert_models/2018_11_23/multi_cased_L-12_H-768_A-12.zip'
download(url)
```

#### BERT Config
- attention_probs_dropout_prob': attention의 dropout 비율
- directionality': 'bidi',
- hidden_act': encoder와 pooler layer를 위한 비선형 activation function
- hidden_dropout_prob': 모든 FC(embeddings, encoder, pooler) layer의 dropout 비율
- hidden_size': encoder layers 와 pooler layer 의 크기
- initializer_range': 0.02,
- intermediate_size': Transformer encoder의 intermediate(feed-forward) layer의 크기
- max_position_embeddings': 이 모델에서 사용할 최대 문장 길이
- num_attention_heads': Transformer encoder안의 각 attention layer의 attention head의 개수
- num_hidden_layers': Transformer encoder의 hidden layer 개수
- pooler_fc_size': 768,
- pooler_num_attention_heads': 12,
- pooler_num_fc_layers': 3,
- pooler_size_per_head': 128,
- pooler_type': 'first_token_transform',
- type_vocab_size': BertModel에 전달되는 token_type_ids의 vocabulary 크기
- vocab_size': 119547

#### KorQuAD : 한국어 질문 답변 Dataset

[KorQuAD](https://korquad.github.io/)

- KorQuAD 2.1의 전체 데이터는 47,957 개의 Wikipedia article에 대해 102,960 개의 질의응답 쌍
- Training set 83,486 개, Dev set 10,165 개의 질의응답쌍으로 구분
- KorQuAD 2.0 데이터 중 HTML 태그의 속성이 완벽하게 제거되지 않은 오류를 수정

```
## KorQuAD

train = 'https://github.com/korquad/korquad.github.io/tree/master/dataset/KorQuAD_2.1/train'
dev = 'https://github.com/korquad/korquad.github.io/tree/master/dataset/KorQuAD_2.1/dev'
```

```
from bs4 import BeautifulSoup as bs

def listFiles(url, ext='zip'):
    page = requests.get(url).text
    soup = bs(page, 'html.parser')
    return [(url + '/' + node.get('href').split('/')[-1]).replace('tree', 'raw') for node in soup.find_all('a') if node.get('href').endswith(ext)]

train_files = listFiles(train)
train_files
```

```
dev_files = listFiles(dev)

for t in train_files:
  print(t)
  download(t, savedir=SAVE_DIR+'train/')

for d in dev_files:
  print(d)
  download(d, savedir=SAVE_DIR+'dev/')
```
